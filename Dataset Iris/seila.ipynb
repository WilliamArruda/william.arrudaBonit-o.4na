```python
import numpy as np
import pandas as pd
from sklearn import datasets

# Carregando o dataset Iris
iris = datasets.load_iris()

# Extraindo os dados e criando um DataFrame
X = iris.data  # Características (features)
y = iris.target  # Classes (target)
feature_names = iris.feature_names
target_names = iris.target_names

# Criando um DataFrame para melhor visualização
df = pd.DataFrame(X, columns=feature_names)
df['species'] = [target_names[i] for i in y]

# Visualizando as primeiras linhas do dataset
print("Primeiras linhas do dataset Iris:")
print(df.head())
```

```python
# Estatísticas descritivas
print("Estatísticas descritivas do dataset Iris:")
print(df.describe())
```

```python
#Verificar a Distribuição das Classes

import matplotlib.pyplot as plt

# Distribuição das classes
plt.figure(figsize=(8, 6))
df['species'].value_counts().plot(kind='bar')
plt.title('Distribuição das Classes no Dataset Iris')
plt.xlabel('Espécies')
plt.ylabel('Número de Amostras')
plt.xticks(rotation=0)
plt.show()
```

```python
import seaborn as sns

# Visualizando a relação entre as características
sns.pairplot(df, hue="species", markers=["o", "s", "D"])
plt.title("Relação entre as Características das Flores")
plt.show()
```
